```python
# Instalar o Matplotlib no Colab (caso ainda não esteja instalado)
!pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do CSV (substitua pelo caminho do seu arquivo CSV)
# Se o arquivo estiver no seu Google Drive, monte o Drive primeiro:
# from google.colab import drive
# drive.mount('/content/drive')

# Substitua '/content/drive/My Drive/pokemon_combined.csv' pelo caminho correto
pokemon_data = pd.read_csv('/content/sample_data/pokemon_combined.csv')

# Distribuição de HP
plt.figure(figsize=(10, 6))
plt.hist(pokemon_data['HP'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de HP entre os Pokémon')
plt.xlabel('HP')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()

# Gráfico de Dispersão de Ataque vs Defesa
plt.figure(figsize=(10, 6))
plt.scatter(pokemon_data['Attack'], pokemon_data['Defense'], color='salmon', edgecolor='black')
plt.title('Ataque vs Defesa')
plt.xlabel('Ataque')
plt.ylabel('Defesa')
plt.grid(True)
plt.show()

# Gráfico de Dispersão de Peso vs Altura
plt.figure(figsize=(10, 6))
plt.scatter(pokemon_data['Weight'], pokemon_data['Height'], color='lightgreen', edgecolor='black')
plt.title('Peso vs Altura')
plt.xlabel('Peso (kg)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()

# Distribuição dos Tipos de Pokémon
plt.figure(figsize=(12, 8))
pokemon_data['Type'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribuição dos Tipos de Pokémon')
plt.xlabel('Tipo')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
